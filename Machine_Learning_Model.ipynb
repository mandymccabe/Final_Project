```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#take input from provisional database (csv)
#df = pd.read_csv(Path('./Resources/all_responses_coded.csv'))
# NEED TO UPDATE URL
# Load the csv file from GitHub
url = 'https://raw.githubusercontent.com/mandymccabe/Final_Project/janet_branch/Data/Archive/TrustInNewsOriginal.csv'
df = pd.read_csv(url, index_col=0)
#df.head()
df.dtypes
```

```python
df.head()
#df.rename(columns={"A1": "Very_Conservative", "A2": "Conservative", "A3": "Moderate", "A4": "Liberal", "A5": "Very_Liberal", "A6":"Great_Deal_Trust", "A7":"Fair_Amount_Trust", "A8": "Not_Very_Much_Trust", "A9": "None_At_All"})
```

```python
#drop columns we don't want to use to predict and those that do not have an int value
df = df.drop(["RespondentID", "A21", "StartDate", "EndDate"], axis=1)
df.head()
```

```python
#start the test with trying to predict whether a person will be "Very conservative" based on the input values
y = df["A1"]
X = df.drop(columns="A1")
```

```python
#create the training model and split into training and testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
   y, random_state=1, stratify=y)
X_train.shape
```

```python
#create the logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
   max_iter=200,
   random_state=1)
```

```python
#train the data with the training model
classifier.fit(X_train, y_train)
```

```python
#use the testing set to create predictions
y_pred = classifier.predict(X_test)
#Predict_DF is our output-predicting, based on the input columns, whether a user is very conservative-labels yes(1) or no(0)
Predict_DF = pd.DataFrame(y_pred, columns=["Prediction"])
#add the prediction df to the actual dataframe to compare output labels to actual answers in A1
Predict_ActualDF = pd.concat([Predict_DF, df], axis=1)
Predict_ActualDF.head(25)
```

```python
#Assess Accuracy
from sklearn.metrics import confusion_matrix, classification_report
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
#True Positives=242
#False Positives=3
#False Negatives=0
#True Negatives=11

#Calculate precision recall f1
report = classification_report(y_test, y_pred)
print(report)
```
